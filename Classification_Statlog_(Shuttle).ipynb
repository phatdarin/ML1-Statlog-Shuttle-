```python
import pandas as pd
train = pd.read_csv("shuttle.trn", sep=" ", header=None)
test = pd.read_csv("shuttle.tst", sep=" ", header=None) #shuttle.tstเปนtext fileอยู่แล้ว ไม่ต้องแตก
```

```python
cols = [f"feature_{i}" for i in range(1,10)] + ["label"]
cols
```

```python
train.columns = cols
test.columns = cols #เอาชื่อcolไปใส่ในdata
```

```python
test.describe()
```

# Preparing

```python
label_map = {
    1: "Rad Flow",
    2: "Fpv Close",
    3: "Fpv Open",
    4: "High",
    5: "Bypass",
    6: "Bpv Close",
    7: "Bpv Open"
}
```

```python
train["label_name"] = train["label"].map(label_map)
test["label_name"] = test["label"].map(label_map)
```

```python
test.head()
```

```python
train["anomaly"] = train["label"].apply(lambda x: 0 if x == 1 else 1) #แปลงเป็น anomaly detection
test["anomaly"] = test["label"].apply(lambda x: 0 if x == 1 else 1) #.apply(lambda x: ...)ใช้กับlabelทุกแถว
```

```python
X_train = train.iloc[:, :-2]
y_train = train["anomaly"]

X_test = test.iloc[:, :-2]
y_test = test["anomaly"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() #ทำให้ค่าอยู่ใน scale เดียวกัน

X_train = scaler.fit_transform(X_train) #เรียนรู้ + แปลง ใช้แค่กับ train
X_test = scaler.transform(X_test) #แปลงอย่างเดียว
```

# Model

**Logistic** **regression**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
y_pred_lr[ :10] #0=normal, 1= anomaly
```

```python
compare_lr = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_lr})
print(compare_lr.head())
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) #ลดจาก9 features เปน 2 features
X_pca = pca.fit_transform(X_train)#แปลง data ให้เหลือ 2 มิติ

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y_train, cmap='coolwarm', s=5) #c=y_train = สีตาม label
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Anomaly Detection Visualization (PCA)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
colors = ['blue' if y == 0 else 'red' for y in y_train] #กำหนดสีเอง
plt.scatter(X_pca[:,0], X_pca[:,1], c=colors, s=5)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Anomaly Detection Visualization (PCA)")
plt.show()
```

```python
from sklearn.manifold import TSNE #เห็น cluster ชัดกว่าแต่ช้ามาก เกือบหลับ

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_train)

plt.figure(figsize=(8,6))
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y_train, cmap='coolwarm', s=5)

plt.title("t-SNE Visualization")
plt.show()
```

## Predict with no Label

```python
A_train = train.drop(columns=['label','label_name','anomaly']) #targetเดิม, text, targetใหม่
B_train = train['anomaly'] #เลือกcolumn anomalyเพื่อทำนาย

A_test = test.drop(columns=['label','label_name','anomaly'])
B_test = test['anomaly']
```

**Logistic Regression**

```python
LR = LogisticRegression()
LR.fit(A_train, B_train)
```

```python
B_pred_LR = LR.predict(A_test)
B_pred_LR[ :10]
```

```python
compare_LR = pd.DataFrame({'Actual': B_test, 'Predicted': B_pred_LR})
print(compare_LR.head())
```

**Random Forest**

```python
RF = RandomForestClassifier()
RF.fit(A_train, B_train)
```

```python
B_pred_RF = RF.predict(A_test)
B_pred_RF[:10]
```

```python
compare_RF = pd.DataFrame({'Actual': B_test, 'Predicted': B_pred_RF})
print(compare_RF.head())
```

**SVM**

```python
B_pred_SVM = SVM.predict(A_test)
B_pred_SVM[:10]
```

```python
compare_SVM = pd.DataFrame({'Actual': B_test, 'Predicted': B_pred_SVM})
print(compare_SVM.head())
```

# **Models Comparing**

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def evaluate(model, name):
    y_pred = model.predict(A_test)

    print(f"=== {name} ===")
    print("Accuracy:", accuracy_score(B_test, y_pred))
    print("\nClassification Report:")
    print(classification_report(B_test, y_pred))

    print("Confusion Matrix:")
    print(confusion_matrix(B_test, y_pred))
    print("\n")
```

```python
evaluate(LR, "Logistic Regression")
evaluate(RF, "Random Forest")
evaluate(SVM, "SVM")
```

```python
from sklearn.metrics import roc_curve, auc

# Logistic Regression
y_score_lr = LR.predict_proba(A_test)[:,1]
fpr_lr, tpr_lr, _ = roc_curve(B_test, y_score_lr)
auc_lr = auc(fpr_lr, tpr_lr)

# Random Forest
y_score_rf = RF.predict_proba(A_test)[:,1]
fpr_rf, tpr_rf, _ = roc_curve(B_test, y_score_rf)
auc_rf = auc(fpr_rf, tpr_rf)

# SVM
y_score_svm = SVM.predict_proba(A_test)[:,1] #Modelต้องเปน SVM = SVC(probabilty=True)
fpr_svm, tpr_svm, _ = roc_curve(B_test, y_score_svm)
auc_svm = auc(fpr_svm, tpr_svm)


plt.figure(figsize=(8,6))
plt.plot(fpr_lr, tpr_lr, label=f"LR (AUC={auc_lr:.2f})")
plt.plot(fpr_rf, tpr_rf, label=f"RF (AUC={auc_rf:.2f})")
plt.plot(fpr_svm, tpr_svm, label=f"SVM (AUC={auc_svm:.2f})")
plt.plot([0,1],[0,1],'k--')  # เส้นสุ่ม
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
